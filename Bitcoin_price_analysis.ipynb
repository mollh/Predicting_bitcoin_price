```python
# Reviewed the heatmap and correlations; other than cryptocurrencies, Google search is most correlated, next is DOW, then NASDAQ, then S&P
#Train and test to make Linear regression for bitcoin price predictions according to the features' values

# Identifying X and Y.
y = df['BTC_Price']

# stocklist = ['Gsearch', 'BTCCash', 'LTC', 'XRP', 'DOW_Price', 'NASDAQ_Price', 'US_BCI', 'US_CCI', 'US30TB']
# Highest value coeff features above 0 r2 = 0.9702137487167897
# Gsearch + crypto_cols + stock_cols + oecd_cols + bonds_cols r2 = 0.9716000479963534
# Gsearch + crypto_cols + stock_cols + oecd_cols r2 = 0.9699264502398567
# Gsearch + crypto_cols + stock_cols r2 = 0.9700646410892597
# Gsearch + crypto_cols r2 = 0.9671119046609067
# oecd_cols r2 = 0.4433119429670074
# bonds_cols r2 = 0.6749213060216392
# forex_cols r2 = 0.03184947870887689
# stock_cols r2 = 0.8080708280820792
# crypto_cols r2 = 0.9014054670808523
# stocklist = ['Gsearch'] r2 = 0.937184878685592
# stocklist = ['US_CCI'] r2 = 0.19184675963395847
# stocklist = ['US_BCI'] r2 = 0.3788484775500748
# stocklist = ['DOW_Price', 'NASDAQ_Price'] r2 = 0.6871090645808486
X = df.drop(["BTC_Price", "Date"],axis=1,inplace=False) # r2 = 0.9732381421490199
#X = df[stocklist]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5000)
```

```python
from sklearn import linear_model
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn import metrics
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
import math
```

```python
# Instantiate one model of each kind with the default settings.

reg = linear_model.Lasso(alpha = 0.1)

ln = linear_model.LinearRegression()

# Train the model on the entire data set.
reg.fit(X_train, y_train)

# Predict the response values for the observations in X ("test the model").
y_pred = reg.predict(X_test)
```

```python
print("train set shape: %s, %s" %(X_train.shape[0],y_train.shape[0]) )
print("test set shape: %s, %s" %(X_test.shape[0],y_train.shape[0]) )
print("pred set shape: %s" %(y_pred.shape[0]) )
```

```python
# Real vs predicted demand for X_test
mse = mean_squared_error(y_test, y_pred)
print (mse)
print (math.sqrt(mse))
```

```python
mae = mean_absolute_error(y_test, y_pred)
print (mae)
```

```python
r2 = r2_score(y_test,y_pred)
print (r2)
```

```python
# calculating coefficients
from pandas import Series, DataFrame

coeff = DataFrame(X_train.columns)

coeff['Coefficient Estimate'] = Series(reg.coef_)
# coeff['Coefficient Estimate']= coeff['Coefficient Estimate'].apply(lambda x: abs(x))
coeff.sort_values('Coefficient Estimate',ascending = False)
```
